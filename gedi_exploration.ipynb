```python
import h5py
import plotly
import geopandas as gpd
import pandas as pd

gedi_swath = h5py.File("GEDI02_B_2019125140338_O02232_T05437_02_001_01.h5", "r")

gdf = gpd.GeoDataFrame()
df = pd.DataFrame()
```

```python
beam = gedi_swath["BEAM0000"]
print(beam)

lats = beam["geolocation"]["latitude_bin0"]
lons = beam["geolocation"]["longitude_bin0"]
pai = beam["pai"]

df = pd.DataFrame({'pai': pai,
           'height':beam['rh100'],
           'beam':beam['beam'],
           'Latitude':lats,
           'Longitude':lons})
```
